```python
import pandas as pd
import yfinance as yf
from pandas_datareader import data as pdr

def fetch_nasdaq_tickers():
    """
    Fetches a complete list of NASDAQ tickers using pandas_datareader.
    """
    nasdaq_tickers = pdr.get_nasdaq_symbols()
    return nasdaq_tickers.index.tolist()

def fetch_data_for_range(tickers, start_date, end_date):
    """
    Downloads historical data for each ticker and checks for trading activity within a specific date range.
    Returns a DataFrame with the results.
    """
    active_tickers = []
    for ticker in tickers:  # Limit to a small number for testing; use multiprocessing or batching for full run
        try:
            # Download data for the specified date range
            data = yf.download(ticker, start=start_date, end=end_date)
            if not data.empty:
                active_tickers.append({
                    'Ticker': ticker,
                    'Total Volume': data['Volume'].sum(),
                    'Average Close': data['Close'].mean()
                })
        except Exception as e:
            print(f"Failed to fetch data for {ticker}: {e}")
    return pd.DataFrame(active_tickers)

if __name__ == "__main__":
    tickers = fetch_nasdaq_tickers()
    # Example date range from April 1, 2024, to May 5, 2024
    active_tickers_df = fetch_data_for_range(tickers[:100], '2024-04-01', '2024-05-05')  # Use a slice for testing
    print(active_tickers_df.head())  # Display the first few entries

    # Save the DataFrame to a CSV file
    active_tickers_df.to_csv('nasdaq_tickers_april_to_may_2024.csv', index=False)
```
